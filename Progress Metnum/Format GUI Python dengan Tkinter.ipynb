```python
import tkinter 

window = tkinter.Tk()

 # Title of the windows 
window.title("Metode Numerik Praktek Python/MATLAB")

# Create a label with pack to show the object in the gui windows
perbandingan = tkinter.Label(window, text = "Perbandingan Metode Trapesium dengan Metode Simpson 1/3" , font=("Times New Roman" , 14))
trapesium = tkinter.Label(window, text = "Metode Trapesium" , font=("Times New Roman" , 12))
simpson_sepertiga = tkinter.Label(window, text = "Metode Simpson 1/3" , font=("Times New Roman" , 12))

# Grid each label in each position

perbandingan.grid(row = 0, column = 2 , pady = 2)
trapesium.grid(row = 2, column = 1 , pady = 2)
simpson_sepertiga.grid(row = 2, column = 4 , pady = 2)

# Make a & b Label for input the variable of Numerical Integration
a_trape = tkinter.Label(window, text = "a" , font=("Times New Roman" , 12))
b_trape = tkinter.Label(window, text = "b" , font=("Times New Roman" , 12))
error_trape = tkinter.Label(window , text = "error" , font=("Times New Roman" , 12))
a_simpson = tkinter.Label(window, text = "a" , font=("Times New Roman" , 12))
b_simpson = tkinter.Label(window, text = "b" , font=("Times New Roman" , 12))
error_simpson = tkinter.Label(window, text = "error" , font=("Times New Roman" , 12))

a_trape.grid(row = 4 , column = 0, pady = 2)
b_trape.grid(row = 5 , column = 0, pady = 2)
error_trape.grid(row = 6 , column = 0 , pady = 2)
a_simpson.grid(row = 4 , column = 3 , pady = 2)
b_simpson.grid(row = 5, column = 3 , pady = 2)
error_simpson.grid(row = 6, column = 3 , pady = 2)

# Membuat textbox dan memposisikan untuk input angkanya
entry_a_trape = tkinter.Entry(window, width = 10)
entry_b_trape = tkinter.Entry(window, width = 10)
entry_error_trape = tkinter.Entry(window , width = 10)
entry_a_simpson = tkinter.Entry(window, width = 10)
entry_b_simpson = tkinter.Entry(window, width = 10)
entry_error_simpson = tkinter.Entry(window , width = 10)

entry_a_trape.grid(row = 4, column = 1 , pady = 2)
entry_b_trape.grid(row = 5, column = 1 , pady = 2)
entry_error_trape.grid(row = 6 , column = 1 , pady = 2)
entry_a_simpson.grid(row = 4, column = 4 , pady = 2)
entry_b_simpson.grid(row = 5, column = 4 , pady = 2)
entry_error_simpson.grid(row = 6 , column = 4 , pady = 2)

# Membuat button dan memposisikan untuk di event handling hasil hitungannya
button_trape = tkinter.Button(window , text = "Calculate Trapesium" , font=("Times New Roman" , 12))
button_simpson = tkinter.Button(window , text = "Calculate Simpson" , font=("Times New Roman" , 12))

button_trape.grid(row = 7, column = 1 , pady = 2)
button_simpson.grid(row = 7 , column = 4 , pady = 2)

window.mainloop()
```

```python
# Cara kedua GUI
from tkinter import *

window = Tk()

# Title of the windows 
window.title("Metode Numerik Praktek Python/MATLAB")
window.geometry("400x200")

GUI_Metnum = PanedWindow(window, orient=VERTICAL)
#GUI_Metnum.pack(fill=BOTH, expand=1)
GUI_Metnum.pack(side = "top" , fill = "y" , expand = False)

TITLE = Label(GUI_Metnum, text="Perbandingan antara Metode Trapesium dan Metode Simpson")
GUI_Metnum.add(TITLE)

Perbandingan = PanedWindow(GUI_Metnum, orient=HORIZONTAL)
GUI_Metnum.add(Perbandingan)

Metode_Trapesium = Label(Perbandingan, text="Metode Trapesium")
a_trape = tkinter.Label(Perbandingan, text = "a" , font=("Times New Roman" , 12))
b_trape = tkinter.Label(Perbandingan, text = "b" , font=("Times New Roman" , 12))
Perbandingan.add(Metode_Trapesium)
Perbandingan.add(a_trape)
Perbandingan.add(b_trape)

Metode_Simpson_sepertiga = Label(Perbandingan, text="Metode Simpson 1/3")
Perbandingan.add(Metode_Simpson_sepertiga)

mainloop()
```
